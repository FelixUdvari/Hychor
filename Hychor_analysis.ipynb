```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('df.csv', encoding='latin1')
```

```python
df
```

```python
df.replace('Not disclosed/Private investment', np.nan, inplace=True)
df.replace('Not Disclosed', np.nan, inplace=True)
df
```

```python
#funding/investment descriptive statistics
df['funding_USD_M'] = pd.to_numeric( df['Funding/Investment Million USD Conversion'], errors='coerce' )
total_funding = df['funding_USD_M'].sum()
print (f"The sum of disclosed and estimated funds/investments of the 27 projects: {total_funding} million USD. Not accounting for undisclosed/private investment.")
```

```python
#funding/investment descriptive statistics

median_funding = df['funding_USD_M'].median()
median_funding
```

```python
mean_funding = df['funding_USD_M'].mean()
mean_funding
```

```python
#total funding across 27 projects
total_funding = df['funding_USD_M'].sum()
total_funding
```

```python
#year operational earliest
year_operational_min = df['Operational Year (Actual/Planned)'].astype(float)
year_operational_min = year_operational_min.min()
year_operational_min
```

```python
#year operational latest
year_operational_max = df['Operational Year (Actual/Planned)'].astype(float)
year_operational_max = year_operational_max.max()
year_operational_max
```

```python
#Number of projects by geographical region/continent
projects_by_region = df['Region'].value_counts()
projects_by_region = projects_by_region.sort_values(ascending=False)
projects_by_region_df = projects_by_region.reset_index()
projects_by_region_df.columns = ['Region', 'Projects']
projects_by_region_df["Project Share (%)"] = (
    projects_by_region_df["Projects"] / projects_by_region_df["Projects"].sum()
) * 100
projects_by_region_df["Project Share (%)"] = projects_by_region_df["Project Share (%)"].round(1)
projects_by_region_df = projects_by_region_df.set_index('Region')
projects_by_region_df
```

```python
#Funding in million USD by region/continent
funding_by_region = df.groupby('Region')['funding_USD_M'].sum()
funding_by_region = funding_by_region.sort_values(ascending=False)
funding_by_region_df = funding_by_region.reset_index().round()
funding_by_region_df.columns = ['Region', 'Funding (M USD)']
funding_by_region_df["Market share (%)"] = (
    funding_by_region_df["Funding (M USD)"] / funding_by_region_df["Funding (M USD)"].sum()
) * 100
funding_by_region_df["Market share (%)"] = funding_by_region_df["Market share (%)"].round(1)
funding_by_region_df = funding_by_region_df.set_index('Region')
funding_by_region_df
```

```python
#Majority Project Status by region
status_counts = df.groupby(["Region", "Status"]).size()
most_common_status = status_counts.groupby(level=0).idxmax().apply(lambda x: x[1])
status_by_region = most_common_status.to_frame(name="Majority Project Status")
status_by_region.index.name = "Region"

status_by_region
```

```python
#majority Technology Type by region
df["Technology Type"] = df["Technology Type"].fillna("Not Disclosed")
technology_counts = df.groupby(["Region", "Technology Type"]).size()
most_common_technology = technology_counts.groupby(level=0).idxmax().apply(lambda x: x[1])
technology_by_region = most_common_technology.to_frame(name="Majority Technology Type")
technology_by_region.index.name = "Region"

technology_by_region
```

```python
#majority Processing Method by region
df["Processing Method"] = df["Processing Method"].fillna("Not Disclosed")
processing_counts = df.groupby(["Region", "Processing Method"]).size()
most_common_processing = processing_counts.groupby(level=0).idxmax().apply(lambda x: x[1])
processing_by_region = most_common_processing.to_frame(name="Majority Processing Method")
processing_by_region.index.name = "Region"

processing_by_region
```

```python
#most common Power Source by region
df["Power Source"] = df["Power Source"].fillna("Not Disclosed")
power_counts = df.groupby(["Region", "Power Source"]).size()
most_common_power = power_counts.groupby(level=0).idxmax().apply(lambda x: x[1])
power_by_region = most_common_power.to_frame(name="Majority Power Source")
power_by_region.index.name = "Region"

power_by_region
```

```python
#mean operational year by region
df['Operational_Year_Num'] = pd.to_numeric(df['Operational Year (Actual/Planned)'], errors='coerce')
mean_operational_by_region = df.groupby('Region')['Operational_Year_Num'].mean().round().astype(int)
mean_operational_by_region.sort_values(ascending=False)

mean_operational_by_region_df = mean_operational_by_region.reset_index()
mean_operational_by_region_df.columns = ['Region', 'Mean Operational Year']
year_by_region = mean_operational_by_region.to_frame(name="Mean Operational Year")
year_by_region.index.name = "Region"
year_by_region
```

```python
# combined dataframe by region
region_df = pd.concat([funding_by_region_df, projects_by_region_df, status_by_region, processing_by_region, power_by_region, year_by_region], axis=1, join="inner")
region_df
```

```python
region_df.to_csv(r'C:/Users/udvar/Dropbox/Hychor/Tables/regions.csv')
```

```python
#Number of projects by  country
projects_by_country = df['Country'].value_counts()
projects_by_country = projects_by_country.sort_values(ascending=False)
projects_by_country_df = projects_by_country.reset_index()
projects_by_country_df.columns = ['Country', 'Projects']
projects_by_country_df["Project Share (%)"] = (
    projects_by_country_df["Projects"] / projects_by_country_df["Projects"].sum()
) * 100
projects_by_country_df["Project Share (%)"] = projects_by_country_df["Project Share (%)"].round(1)
projects_by_country_df = projects_by_country_df.set_index('Country')
projects_by_country_df
```

```python
#funding_by_country in million USD
funding_by_country = df.groupby('Country')['funding_USD_M'].sum()
funding_by_country = funding_by_country.sort_values(ascending=False)
funding_by_country_df = funding_by_country.reset_index().round()
funding_by_country_df.columns = ['Country', 'Funding (M USD)']

funding_by_country_df["Market Share (%)"] = (
    funding_by_country_df["Funding (M USD)"] / funding_by_country_df["Funding (M USD)"].sum()
) * 100
funding_by_country_df["Market Share (%)"] = funding_by_country_df["Market Share (%)"].round(3)
funding_by_country_df = funding_by_country_df.set_index('Country')
funding_by_country_df

funding_by_country_df
```

```python
#Majority Project Status by country
status_by_country = df.groupby(["Country", "Status"]).size()
status_by_country = status_by_country.groupby(level=0).idxmax().apply(lambda x: x[1])
status_by_country = status_by_country.to_frame(name="Majority Project Status")
status_by_country.index.name = "Country"

status_by_country
```

```python
#most common Power Source by country
df["Power Source"] = df["Power Source"].fillna("Not Disclosed")
power_by_country = df.groupby(["Country", "Power Source"]).size()
power_by_country = power_by_country.groupby(level=0).idxmax().apply(lambda x: x[1])
power_by_country = power_by_country.to_frame(name="Majority Power Source")
power_by_country.index.name = "Country"

power_by_country
```

```python
#mean operational year by country
mean_operational_by_country = df.groupby('Country')['Operational_Year_Num'].mean().round()
mean_operational_by_country.sort_values(ascending=False)
#creating a dataframe of mean operational year by country
year_by_country = mean_operational_by_country.reset_index()
year_by_country.columns = ['Country', 'Mean Operational Year']
year_by_country.set_index("Country", inplace=True)
year_by_country
```

```python
#majority Processing Method by country
df["Processing Method"] = df["Processing Method"].fillna("Not Disclosed")
processing_country = df.groupby(["Country", "Processing Method"]).size()
most_common_processing = processing_country.groupby(level=0).idxmax().apply(lambda x: x[1])
processing_by_country = most_common_processing.to_frame(name="Majority Processing Method")
processing_by_country.index.name = "Country"

processing_by_country
```

```python
# combined dataframe by region
countries_df = pd.concat([funding_by_country_df, projects_by_country_df, status_by_country, processing_by_country, power_by_country, year_by_country], axis=1, join="inner")
countries_df
```

```python
#bar chart by region
stacked_df = region_df[['Market share (%)','Project Share (%)']].reset_index()

regions = stacked_df['Region']
market = stacked_df['Market share (%)']
project = stacked_df['Project Share (%)']

plt.rcParams.update({
    'font.family': 'Times New Roman',  # font
    'font.size': 14,                   # size
    'text.color': 'black',             # text color (fully opaque)
    'axes.labelcolor': 'black',        # axis labels
    'xtick.color': 'black',            # x-axis tick labels
    'ytick.color': 'black',            # y-axis tick labels
    'legend.facecolor': 'white',       # legend background fully opaque
    'legend.edgecolor': 'black'        # legend border
})

x = np.arange(len(regions))   # positions
width = 0.35                  # bar width

plt.figure(figsize=(10,6))

plt.bar(x - width/2, market,  width, label='Market Share (%)',  color='#FF7F7F')
plt.bar(x + width/2, project, width, label='Project Share (%)', color='#88E788')

plt.xticks(x, regions)
plt.ylabel("Market Allocation (%)", )
plt.xlabel("Region")
plt.title("SHI Market Share & Project Share by Region")
plt.legend()

plt.ylim(0, 60)   # you asked for y-axis going up to 100%

plt.tight_layout()

plt.savefig(r'C:/Users/udvar/Dropbox/Hychor/Tables/Market Share vs Project Share by Region.png', dpi=600)
            
plt.show()
```
